```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create a dictionary of service and initialization time data
data = {
    'Service': ['dbaas-server', 'a1mediator', 'alarmmanager', 'appmgr', 'e2mgr', 
                'o1mediator', 'rtmgr', 'submgr', 'vespamgr', 'RIC-Man.'],
    'Initialization(s)': [28.598, 10.124, 1.842, 3.679, 10.126, 1.686, 15.145, 15.102, 1.763, 50.963],
    'Std_dev': [4.632, 0.014, 0.165, 0.136, 0.008, 0.242, 0.009, 0.002, 0.272, 0.031]
}

```

```python
fig, ax = plt.subplots(figsize=(5, 2))

# Convert dictionary to pandas DataFrame
df = pd.DataFrame(data)

# Create the gray color palette
gray_palette = sns.color_palette("gray", n_colors=1)

# Create the bar plot using Seaborn with the gray color palette
ax = sns.barplot(x="Initialization(s)", y="Service", data=df,edgecolor='black', color='#6495ED', xerr=df['Std_dev'])
ax.xaxis.grid(True)

# Put the axis grid on the background
ax.set_axisbelow(True)

# Add values on top of the bars, offset by standard deviation
for i, (v, sd) in enumerate(zip(df['Initialization(s)'], df['Std_dev'])):
    ax.text(v + sd + 0.2, i + 0.2, "{:.2f}".format(v), fontsize=10)

# Set the plot title and labels
ax.set_xlabel("Initialization Time (s)")
ax.set_ylabel("Near-RT RIC Service")

# Show the plot
plt.savefig('out/RIC-Man-Initialization-Times.pdf', bbox_inches='tight')
plt.savefig('out/RIC-Man-Initialization-Times.png', dpi=300, bbox_inches='tight')
plt.show()
print("Current figure size:", fig.get_size_inches())
```
